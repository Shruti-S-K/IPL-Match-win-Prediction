## Match status in 2nd innings (Feature Engineering)

```python
delv = matches.merge(delv,on='match_id')
delv.head(3)
```

```python
delv = delv[delv['inning'] == 2]
delv
```

## Add Current Score

```python
delv['curr_score'] = delv.groupby('match_id').cumsum()['total_runs_y']
delv.head()
```

```python
delv['runs_left'] = delv['total_runs_x'] - delv['curr_score']
delv['runs_left'] = delv['runs_left'].apply(lambda x:x if x>=0 else 0)
delv.head(3)
```

```python
delv['balls_left'] = 120 - ((delv['over']-1)*6 + delv['ball'])
delv.head(3)
```

```python
delv['player_dismissed'] = delv['player_dismissed'].fillna(0)
delv['player_dismissed'] = delv['player_dismissed'].apply(lambda x:x if x == 0 else 1)
delv.head(3)
```

```python
wkts = delv.groupby('match_id').cumsum()['player_dismissed'].values
```

```python
delv['wkts_left'] = 10 - wkts
delv.head(3)
```

```python
delv['crr'] = (delv['curr_score'] * 6)/(120-delv['balls_left'])
delv.head(3)
```

```python
delv['rrr'] = (delv['runs_left']*6)/delv['balls_left']
delv.head(3)
```

### Did the team batting in 2nd innings win or not?

```python
def match_results(dset):
    return 1 if dset['batting_team']==dset['winner'] else 0
delv['results'] = delv.apply(match_results,axis=1)
delv.head(3)
```

```python
final_dset = delv[['batting_team','bowling_team','city','runs_left','balls_left','wkts_left','crr','rrr','results']]
final_dset
```

```python
final_dset.dropna(inplace=True)
final_dset
```

```python
(final_dset['balls_left'] == 0).sum()
```

```python
final_dset = final_dset[final_dset['balls_left']!=0]
final_dset
```

```python
final_dset['batting_team'].unique()
```

```python
final_dset['batting_team'] = final_dset['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
final_dset['bowling_team'] = final_dset['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

final_dset['batting_team'] = final_dset['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
final_dset['bowling_team'] = final_dset['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
final_dset['batting_team'].unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
final_dset['batting_team'] = enc.fit_transform(final_dset['batting_team'])
final_dset['bowling_team'] = enc.fit_transform(final_dset['bowling_team'])
final_dset['city'] = enc.fit_transform(final_dset['city'])
```

```python
final_dset['batting_team'].unique()
```

```python
X = final_dset.iloc[:,:-1]
y = final_dset.results
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
delv['batting_team'] = delv['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
delv['bowling_team'] = delv['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

delv['batting_team'] = delv['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
delv['bowling_team'] = delv['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
def win_prob_calc(dset,m_id,model):
    match = dset[dset['match_id']==m_id]
    print(match['batting_team'].iloc[0],"vs",match['bowling_team'].iloc[0])
    match = match[(match['ball']==6)]
    final = match[['batting_team','bowling_team','city','runs_left','balls_left','wkts_left','crr','rrr']].dropna()
    final = final[final['balls_left']!=0]
    final['batting_team'] = enc.fit_transform(final['batting_team'])
    final['bowling_team'] = enc.fit_transform(final['bowling_team'])
    final['city'] = enc.fit_transform(final['city'])
    result = model.predict_proba(final)
    
    final['lose'] = np.round(result.T[0]*100,1)
    final['win'] = np.round(result.T[1]*100,1)
    
    final['end_of_over'] = range(1,final.shape[0]+1)
    target = match['total_runs_x'].values[0]
    print("Target:",target)
    
    runs = list(final['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    final['runs_in_over'] = np.array(runs)[:-1] - np.array(new_runs)
    
    wickets = list(final['wkts_left'].values)
    new_wkts = wickets[:]
    new_wkts.insert(0,10)
    wickets.append(0)
    
    w = np.array(wickets)
    nw = np.array(new_wkts)
    
    final['wkts_in_over'] = (nw-w)[0:final.shape[0]]
    final = final[['end_of_over','runs_in_over','wkts_in_over','lose','win']]
    
    print("Win probability for",dset['batting_team'].iloc[0])
    return final, target
```

```python
out_df, target = win_prob_calc(delv,80,model)
```
